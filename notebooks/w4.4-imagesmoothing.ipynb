```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib

matplotlib.style.use('dark_background')
matplotlib.rcParams['figure.figsize'] = (5.0, 5.0)
matplotlib.rcParams['image.cmap'] = 'gray'
matplotlib.rcParams['image.interpolation'] = 'bilinear'
%matplotlib inline
```

```python
DATA_PATH = "D:\Estudos\OpenCV\Computer Vision - Introduction\w4 - Image Enhancement and Filtering\data"

img = cv2.imread(DATA_PATH + "\images\gaussian-noise.png")
plt.figure()
plt.imshow(img[:, :, ::-1]);
```

## Box Blur

```python
dst1 = cv2.blur(img, (3, 3), (-1, -1))
dst2 = cv2.blur(img, (7, 7), (-1, -1))
```

```python
plt.figure(figsize=(10,10))

ax = plt.subplot(131)
plt.imshow(img[:,:,::-1])
ax.axis('off')

ax = plt.subplot(132)
plt.imshow(dst1[:,:,::-1])
ax.axis('off')

ax = plt.subplot(133)
plt.imshow(dst2[:,:,::-1])
ax.axis('off')
plt.show()
```

## Gaussian Blur

```python
dst1 = cv2.GaussianBlur(img, (5,5), 0, 0)
dst2 = cv2.GaussianBlur(img, (25,25), 50, 50)
```

```python
plt.figure(figsize=(10, 10))

ax = plt.subplot(131)
plt.imshow(img[:, :, ::-1])
ax.axis('off')

ax = plt.subplot(132)
plt.imshow(dst1[:, :, ::-1])
ax.axis('off')

ax = plt.subplot(133)
plt.imshow(dst2[:, :, ::-1])
ax.axis('off')
plt.show()
```

## Median Blur

```python
DATA_PATH = "D:\Estudos\OpenCV\Computer Vision - Introduction\w4 - Image Enhancement and Filtering\data"

img = cv2.imread(DATA_PATH + "\images\salt-and-pepper.png")
```

```python
kernelSize = 5

medianBlurred = cv2.medianBlur(img,kernelSize)
```

```python
plt.figure(figsize=(10, 10))

ax = plt.subplot(121)
plt.imshow(img[:, :, ::-1])
ax.axis('off')

ax = plt.subplot(122)
plt.imshow(medianBlurred[:, :, ::-1])
ax.axis('off')
plt.show()
```

## Bilateral Filters

```python
DATA_PATH = "D:\Estudos\OpenCV\Computer Vision - Introduction\w4 - Image Enhancement and Filtering\data"

img = cv2.imread(DATA_PATH + "\images\gaussian-noise.png")
```

```python
dia = 15
sigmaC = 80
sigmaS = 80

result = cv2.bilateralFilter(img, dia, sigmaC, sigmaS)
```

```python
plt.figure(figsize=(10, 10))

ax = plt.subplot(121)
plt.imshow(img[:, :, ::-1])
ax.axis('off')

ax = plt.subplot(122)
plt.imshow(result[:, :, ::-1])
ax.axis('off')
plt.show()
```
