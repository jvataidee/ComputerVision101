## Color Image

```python
elonmusk = f"{DATA_PATH}\\images\\boy.jpg"
elonmusk = cv2.imread(elonmusk, 1)
print(f"Image Dimension y={elonmusk.shape[0]}, x={elonmusk.shape[1]} and chanel={elonmusk.shape[2]}")
```

## Image Channels

```python
plt.imshow(elonmusk);
```

```python
elonRGB = cv2.cvtColor(elonmusk, cv2.COLOR_BGR2RGB)
plt.imshow(elonRGB);
```

```python
plt.figure(figsize=[10,5])
plt.imshow(elonmusk[:,:,::-1]);
```

```python
plt.figure(figsize=[20,5])

plt.subplot(131);plt.imshow(elonmusk[:,:,0]);plt.title("Blue Channel");
plt.subplot(132);plt.imshow(elonmusk[:,:,1]);plt.title("Green Channel");
plt.subplot(133);plt.imshow(elonmusk[:,:,2]);plt.title("Red Channel");
```

## Splitting and Merging Channels

```python
plt.figure(figsize=[20,5])

b, g, r = cv2.split(elonmusk)
plt.subplot(141);plt.imshow(r);plt.title("Red Channel");
plt.subplot(142);plt.imshow(g);plt.title("Green Channel");
plt.subplot(143);plt.imshow(b);plt.title("Blue Channel");
```

```python
elonMerg = cv2.merge((b, g, r))
plt.imshow(elonMerg[:,:,::-1]);plt.title("Merged Output");
```
