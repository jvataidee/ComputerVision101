```python
#Importando as bibliotecas
import cv2
import numpy as np

#image configs
import matplotlib.pyplot as plt
import matplotlib

matplotlib.style.use('dark_background')
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
%matplotlib inline

DATA_PATH = r"D:\Estudos\OpenCV\Computer Vision - Introduction\w7 - Geometric Transformations and Image Features\data"
```

## ORB in OpenCV


```python
img = cv2.imread(DATA_PATH + "\\images\\book.jpeg")
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(121)
plt.title("Original")
plt.imshow(img[:, :, ::-1])
plt.subplot(122)
plt.title("Gray scale")
plt.imshow(imgGray)
plt.show()
```

```python
orb = cv2.ORB_create()

keypoints = orb.detect(imgGray, None)
```

```python
keypoints, descriptors = orb.compute(imgGray, keypoints)
```

```python
img2 = cv2.drawKeypoints(img, keypoints, None, color = (0, 255, 0),
                         flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.title("Original with keypoints")
plt.imshow(img2[:, :, ::-1])
plt.show()
```

```python
orb = cv2.ORB_create(10)

keypoints = orb.detect(imgGray, None)
keypoints, descriptors = orb.compute(imgGray, keypoints)
img2 = cv2.drawKeypoints(img, keypoints, None, color=(0, 255, 0),
                         flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.title("Original with keypoints")
plt.imshow(img2[:, :, ::-1])
plt.show()
```
