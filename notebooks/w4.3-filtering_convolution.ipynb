```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib

matplotlib.style.use('dark_background')
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
matplotlib.rcParams['image.cmap'] = 'gray'
matplotlib.rcParams['image.interpolation'] = 'bilinear'
%matplotlib inline
```

```python
DATA_PATH = "D:\Estudos\OpenCV\Computer Vision - Introduction\w4 - Image Enhancement and Filtering\data"

img = cv2.imread(DATA_PATH+"\images\sample.jpg")
plt.figure()
plt.imshow(img[:,:,::-1]);
```

```python
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), dtype=np.float32)/kernel_size**2
print(kernel)
```

```python
result = cv2.filter2D(img, -1, kernel,
                      (-1, -1), delta = 0,
                      borderType=cv2.BORDER_DEFAULT)
```

```python
plt.figure()

ax = plt.subplot(121)
plt.imshow(img[:,:,::-1])
ax.axis('off')

ax = plt.subplot(122)
plt.imshow(result[:,:,::-1])
ax.axis('off')
plt.show()
```

```python
kernel = np.array([[-1, 0, 1],
                  [-2, 0, 2],
                  [-1, 0, 1]])

result = cv2.filter2D(img, -1, kernel,
                      (-1, -1), delta = 0,
                      borderType=cv2.BORDER_DEFAULT)

plt.figure()

ax = plt.subplot(121)
plt.imshow(img[:,:,::-1])
ax.axis('off')

ax = plt.subplot(122)
plt.imshow(result[:,:,::-1])
ax.axis('off')
plt.show()
```
