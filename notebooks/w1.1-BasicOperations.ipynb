```python
#Importando as bibliotecas
import cv2
import numpy as np

#image configs
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('dark_background')
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
%matplotlib inline
```

```python
DATA_PATH = r"D:\Estudos\OpenCV\Computer Vision - Introduction\w1 - Getting Started\data"
```

## Create New Image

```python
imagePath = f"{DATA_PATH}\\images\\boy.jpg"
image = cv2.imread(imagePath, 1)
plt.imshow(image[:,:,::-1]);
```

```python
imageCopy = image.copy()
```

```python
emptyMatrix = np.zeros(imageCopy.shape, dtype='uint8')
plt.imshow(emptyMatrix);
```

```python
emptyMatrix = 255*np.ones(imageCopy.shape, dtype='uint8')
plt.imshow(emptyMatrix);
```

```python
emptyMatrix = 100*np.ones_like(imageCopy)
plt.imshow(emptyMatrix);
```

## Cropping Image

```python
crop = imageCopy[40:200, 170:320]
plt.subplot(122);plt.imshow(crop[:, :, ::-1]);plt.title("Original Image");
```

## Copying Region to Another

```python
cropH, cropW = crop.shape[:2]

imageCopy[40:40+cropH, 10:10+cropW] = crop
imageCopy[40:40+cropH, 330:330 +cropW] = crop

# Display the output
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original Image");
plt.subplot(122);plt.imshow(imageCopy[...,::-1]);plt.title("Output Image");
```

## Resizing an Image

### Specify W and H

```python
resDownW = 300
resDownH = 200
resDown = cv2.resize(image, (resDownW, resDownH), interpolation = cv2.INTER_LINEAR)
```

```python
resUpW = 600
resUpH = 900
resUp = cv2.resize(image, (resUpW, resUpH), interpolation = cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image[:,:,::-1]);plt.title("Original Image")
plt.subplot(132);plt.imshow(resUp[:,:,::-1]);plt.title("Scaled Up Image")
plt.subplot(133);plt.imshow(resDown[:,:,::-1]);plt.title("Scaled Down Image");
```

### Specify Scaling Factor

```python
scaleDownY = 0.6
scaleDownX = 0.6

resDown = cv2.resize(image, None, fx = scaleDownX, fy = scaleDownY, interpolation=cv2.INTER_LINEAR)
```

```python
scaleUpX = 2
scaleUpY = 2

resUp = cv2.resize(image, None, fx = scaleUpX, fy = scaleUpY, interpolation=cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image[:,:,::-1]);plt.title("Original Image")
plt.subplot(132);plt.imshow(resUp[:,:,::-1]);plt.title("Scaled Up Image")
plt.subplot(133);plt.imshow(resDown[:,:,::-1]);plt.title("Scaled Down Image");
```

## Creating Image Mask

```python
mask1 = np.zeros(image.shape[:2], dtype='uint8')
plt.imshow(mask1);
```

```python
mask1[50:200, 170:320] = 255
plt.imshow(mask1);
```

```python
masked = cv2.bitwise_and(image, image, mask = mask1)
plt.imshow(masked[:,:,::-1])
```

### Mask with intensity or color

```python
mask2 = cv2.inRange(image, (0, 0, 150), (100, 100, 255))
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original Image")
plt.subplot(122);plt.imshow(mask2);plt.title("Masked Image");
```

```python
masked = cv2.bitwise_and(image, image, mask = mask2)
plt.imshow(masked[:,:,::-1]);
```
